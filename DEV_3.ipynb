```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt

# Ensure x and y match in length
x = [1, 2, 3, 4, 5]
b = [0, 0.6, 0.2, 1.5, 1.0]       # matched to x (5 points)
c = [4, 2, 6, 8, 3]               # also matched to x

plt.plot(x, label='x')            # simple line
plt.plot(x, b, 'or', label='b')   # red circles for b
plt.plot(x, c, label='4th Rep')   # default line for c

# Clean up axes
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(-3, 40)
plt.xlabel('Day')
plt.ylabel('Temp')
plt.xticks(range(1, 6))
plt.yticks(range(-3, 21, 3))
plt.legend()
plt.title('My Data Plot')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

a = [1, 2, 3, 4, 5, 6, 7, 8]
b = [0, 0.6, 0.2, 15, 10, 8, 16, 21]
c = [4, 2, 6, 8, 3, 20, 13, 15]

fig = plt.figure(figsize=(10, 10))
sub1 = fig.add_subplot(2, 2, 1)
sub2 = fig.add_subplot(2, 2, 2)
sub3 = fig.add_subplot(2, 2, 3)
sub4 = fig.add_subplot(2, 2, 4)

x = list(range(len(a)))  # [0..7], same length as a, b, c

sub1.plot(x, a, 'sb')
sub1.set_xticks(range(0, 9, 1))
sub1.set_title('1st Rep')

sub2.plot(x, b, 'or')
sub2.set_xticks(range(0, 9, 2))
sub2.set_title('2nd Rep')

sub3.plot(x, b, 'vg')
sub3.set_xticks(range(0, 9, 1))
sub3.set_title('3rd Rep')

sub4.plot(x, c, 'Dm')
sub4.set_yticks(range(0, 24, 2))
sub4.set_title('4th Rep')

plt.tight_layout()
plt.show()
```
